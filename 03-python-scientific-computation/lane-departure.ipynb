```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# Specify path relative to current file.
img_path = "./res/road.jpg"
# Read the original image and convert to the 3 channel BGR color.
img = cv.imread(img_path, cv.IMREAD_COLOR)

# Display original image
img_plot = plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.show() 
```

```python
# Convert to graycsale
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# Blur the image for better edge detection
img_blur = cv.GaussianBlur(img_gray, (3,3), 0)

img_plot = plt.imshow(cv.cvtColor(img_blur, cv.COLOR_BGR2RGB))
plt.show() 
```

```python
img_edges = cv.Canny(img_blur, 100, 200)

img_plot = plt.imshow(cv.cvtColor(img_edges, cv.COLOR_BGR2RGB))
plt.show() 
```
